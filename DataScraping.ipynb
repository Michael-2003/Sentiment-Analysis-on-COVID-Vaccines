```python
import pandas as pd 

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
```

```python
# Create a new instance of the Firefox driver
driver = webdriver.Firefox()

# Go to the Tumblr page
driver.get('https://www.tumblr.com/search/covid%20pandemic%20coronavirus%20covid19/text?t=1300')

# Get the body of the page
body = driver.find_element(By.CSS_SELECTOR, 'body')

# Scroll down
for _ in range(1000):  # adjust this range according to your needs
    time.sleep(0.2)
    body.send_keys(Keys.PAGE_DOWN)
    
  
# Now you can parse the page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml')
articles = soup.find_all('article')

posts = []
        
for article in articles:
    post = {}

    # Author
    author_link = article.find('a', class_="zHhsx BSUG4")
    post['Author'] = author_link.text
       
    # Paragraphs
    paragraphs = article.find_all('p')
    post['Paragraphs'] = [p.text for p in paragraphs]
    posts.append(post)
 
# Don't forget to close the driver
driver.quit()
```
